Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful data visualization library in Python that provides an easy-to-use interface for creating various types of statistical plots. Here are five common plots that you can create using Seaborn and their typical use cases:

1. **Bar Plot**:
   - **Use**: Bar plots are used to visualize the distribution of a categorical variable. They display the count or mean of data in different categories.
   
2. **Histogram**:
   - **Use**: Histograms are used to show the distribution of a single numerical variable. They display the frequency or density of data points within defined intervals (bins).

3. **Scatter Plot**:
   - **Use**: Scatter plots are used to visualize the relationship between two numerical variables. They display individual data points as dots and are useful for identifying patterns, correlations, or outliers.

4. **Box Plot (Box-and-Whisker Plot)**:
   - **Use**: Box plots are used to summarize the distribution of a numerical variable and identify outliers. They show the median, quartiles, and potential outliers in the data.

5. **Heatmap**:
   - **Use**: Heatmaps are used to represent data in a matrix format, where color intensity represents the value of each element. They are commonly used for visualizing correlation matrices, hierarchical clustering, and data matrices.

These are just a few of the many types of plots that Seaborn can create. Seaborn provides a high-level interface to create aesthetically pleasing and informative data visualizations, making it a popular choice for data analysts and scientists.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title("Line Plot of FMRI Data")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create two box plots using Seaborn
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of Age by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")

plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of Fare by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x="price", hue="cut", kde=True, bins=30)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend(title="Cut")
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue
sns.set(style="ticks")
sns.pairplot(iris_data, hue="species")
plt.suptitle("Pair Plot of Iris Dataset")
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm")
plt.title("Flights Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
