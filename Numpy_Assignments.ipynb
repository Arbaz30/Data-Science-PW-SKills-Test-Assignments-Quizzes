Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

**Matplotlib** is a popular data visualization library in Python used for creating static, animated, and interactive plots and charts. It provides a wide range of functionalities to create various types of plots and graphs, making it a valuable tool for data analysis and visualization.

**Matplotlib is used for the following purposes**:

1. **Data Visualization**: It is used to visualize data and display it in a graphical format, which is often more intuitive and easier to interpret than raw data.

2. **Exploratory Data Analysis (EDA)**: Data scientists and analysts use Matplotlib to explore datasets, identify patterns, and gain insights by creating various types of visualizations.

3. **Communicating Results**: Matplotlib is used to create charts and plots for reports, presentations, and publications, making it easier to convey findings to a wider audience.

4. **Model Validation and Testing**: In machine learning and scientific research, Matplotlib is used to visualize model outputs, validation results, and test data.

5. **Customization**: Matplotlib offers extensive customization options, allowing users to control the appearance and styling of plots to meet specific needs.

**Five types of plots that can be created using the Pyplot module of Matplotlib**:

1. **Line Plot**: Used to display data points or trends over time. It is created using `plt.plot()`.

2. **Bar Chart**: Used to represent categorical data. It is created using `plt.bar()` or `plt.barh()`.

3. **Scatter Plot**: Used to show the relationship between two variables by plotting individual data points. It is created using `plt.scatter()`.

4. **Histogram**: Used to visualize the distribution of a dataset. It is created using `plt.hist()`.

5. **Pie Chart**: Used to represent parts of a whole. It is created using `plt.pie()`.

Matplotlib's Pyplot module offers a wide variety of functions and settings to create and customize these and many other types of plots and charts.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, c='blue', marker='o', label='Data Points')

# Add title and labels
plt.title('Scatter Plot of Random Data')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Show the plot
plt.legend()
plt.grid()
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)  # Create a 2x2 grid of subplots, and select the first subplot
plt.plot(x, y1, label='Line 1')
plt.title('Line 1')
plt.legend()

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)  # Select the second subplot
plt.plot(x, y2, label='Line 2')
plt.title('Line 2')
plt.legend()

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)  # Select the third subplot
plt.plot(x, y3, label='Line 3')
plt.title('Line 3')
plt.legend()

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)  # Select the fourth subplot
plt.plot(x, y4, label='Line 4')
plt.title('Line 4')
plt.legend()

# Adjust layout and display the subplots
plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A **bar plot** is a type of data visualization that displays data using rectangular bars or columns. Bar plots are used to compare different categories or groups and represent the data as the height or length of the bars. They are commonly used for displaying categorical data and making comparisons between different entities.

Bar plots are used for the following purposes:

1. **Comparisons**: They are useful for comparing data between different categories or groups, making it easy to identify trends, patterns, or differences.

2. **Ranking**: Bar plots can help rank entities based on a specific attribute, such as profit or sales.

3. **Distribution**: They show the distribution of data within categories, making it clear whether data is skewed or evenly distributed.

4. **Visual Representation**: Bar plots provide a visually appealing way to represent data and communicate information effectively.

Here's how you can create both a bar plot and a horizontal bar plot using the provided data:




```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.title('Bar Plot of Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.xticks(rotation=45)

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='salmon')
plt.title('Horizontal Bar Plot of Company Profits')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')

plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A **box plot**, also known as a box-and-whisker plot, is a data visualization that provides a graphical summary of the distribution of a dataset. It is used to display the following statistical information about a dataset:

- The median (middle value) of the data.
- The first quartile (25th percentile) and the third quartile (75th percentile).
- The range of the data, which is represented by the "box" that spans from the first quartile to the third quartile.
- Outliers, which are data points that fall significantly above or below the rest of the data, displayed as individual points or "whiskers" extending from the box.

Box plots are particularly useful for:

1. **Identifying Skewness**: They help in understanding the skewness (positively or negatively skewed) of the data.

2. **Comparing Distributions**: They allow for easy comparison of distributions between different groups or categories.

3. **Detecting Outliers**: Outliers are visually apparent in box plots, making them useful for detecting and understanding extreme data points.

Here's how you can create a box plot using the provided data:


```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Two Datasets')
plt.ylabel('Values')

plt.show()
```
